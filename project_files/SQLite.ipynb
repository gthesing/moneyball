```python
from sqlalchemy import create_engine, Column, Integer, String
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import Session
import pymysql
import pandas as pd
import warnings
import os
import sqlite3
pymysql.install_as_MySQLdb()
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Resources/baseball.csv")
df.head()
```

```python
baseball_df = df[['Team','League','Year','RS','RA','W','OBP','SLG','BA','Playoffs','RankPlayoffs','OOBP','OSLG']]
baseball_df.head()
```

```python
Base = declarative_base()
class Baseball(Base):
    __tablename__ = 'baseball'
    id = Column(Integer, primary_key=True)
    Team = Column(String(10))
    League = Column(String(5))
    Year = Column(Integer)
    RS = Column(Integer)
    RA = Column(Integer)
    W = Column(Integer)
    OBP = Column(Integer)
    SLG = Column(Integer)
    BA = Column(Integer)
    Playoffs = Column(Integer)
    RankPlayoffs = Column(String(255))
    OOBP = Column(Integer)
    OSLG = Column(Integer)
```

```python
engine = create_engine("sqlite:///baseball.sqlite")
conn = engine.connect()
Base.metadata.create_all(conn)
session = Session(bind=engine)
```

```python
baseball_df.to_sql("baseball", conn, if_exists='append', index=False)
```
